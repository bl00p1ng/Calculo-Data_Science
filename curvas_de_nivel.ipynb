# Diseñando mapas: curvas de nível

## Dependencias

```python
from matplotlib import cm # Para manejar colores
import numpy as np
import matplotlib.pyplot as plt
```

## Gráfica en 3 dimensiones

```python
def z(x, y):
  return np.sin(x) + 2 * np.cos(y)

res = 100

x = np.linspace(-4, 4, num=res)
y = np.linspace(-4, 4, num=res)

x, y = np.meshgrid(x, y)

z = z(x, y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

surf = ax.plot_surface(x, y, z, cmap=cm.cool)
fig.colorbar(surf)
```

## Curvas de nivel

```python
fig2, ax2 = plt.subplots()
level_map = np.linspace(np.min(z), np.max(z), num=res)

cp = ax2.contour(x, y, z, leves=level_map, cmap=cm.cool)
fig2.colorbar(cp)
```
